```python
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url='http://www.espn.com/nhl/statistics/player/_/stat/points/sort/points/year/2015/seasontype/2'
```

```python
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, "lxml")
```

```python
table=soup.find_all('table')[0] #find the first table
```

```python
rows=table.find_all('tr')
```

```python
player_data=[]
```

```python
for row in rows:
  player_list= [td.get_text() for td in row.find_all("td")]
  player_data.append(player_list)
```

```python
df=pd.DataFrame(player_data)
```

```python
df
```

```python
df1=df.dropna(axis=0, thresh=df.shape[1]-3).reset_index(drop=True)
df1
```

```python
df1=df1.drop(df.index[[11,22,33]]).reset_index(drop=True)
df1
```

```python
df1=df1.rename(columns=df1.loc[0,:])
```

```python
df1=df1.drop(df1.index[0]).drop(axis=1, labels=['RK']).reset_index(drop=True)
df1
```
